```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

Generating 10 random numbers between 25 to 35

```python
age=[random.randrange(25,35,1) for ages in range(11)]
ages=sorted(age,reverse=False)
print("Unsorted",age)
print("Sorted",ages)
```

Generating 10 random numbers between 30k to 45k

```python
dev=[random.randrange(30000,45000,1) for ages in range(11)]
devs=sorted(dev,reverse=False)
print("Unsorted",dev)
print("Sorted",devs)
```

PLotting the Line Graph

```python
plt.plot(ages,devs)
plt.xlabel("Age")
plt.ylabel("Salary in $")
plt.show()
```

```python
py_dev=[random.randrange(50000,75000,1) for ages in range(11)]
py_devs=sorted(py_dev,reverse=False)
```

Plotting the Graph

```python
plt.plot(ages,devs,label="All Developers")
plt.plot(ages,py_devs,label="Python Developers")
plt.xlabel("Age")
plt.ylabel("Salary in $")
plt.title("Median Salary in $ with Age")
plt.legend()
plt.show()
```

```python
plt.plot(ages,devs,label="All Developers",color="Blue",linestyle="--",marker="D")
plt.plot(ages,py_devs,label="Python Developers",color="Red",linestyle="-",marker="o")
plt.xlabel("Age")
plt.ylabel("Salary in $")
plt.title("Median Salary in $ with Age")
plt.legend()
plt.show()
```

```python
js_dev=[random.randrange(40000,60000) for js_dev in range(11)]
js_devs=sorted(js_dev,reverse=False)
```

```python
plt.plot(ages,devs,label="All Developers",color="#FF33E9",linestyle="--",marker="D",)
plt.plot(ages,py_devs,label="Python Developers",color="#3344FF",linestyle="-.",marker="o",)
plt.plot(ages,js_devs,label="Java Script Developers",color="#FF3355",linestyle=":",marker="x",)
plt.xlabel("Age")
plt.ylabel("Salary in $")
plt.title("Median Salary in $ with Age")
plt.legend()
plt.show()
```

```python
plt.plot(ages,devs,label="All Developers",color="#FF33E9",linestyle="--",marker="D",linewidth=4)
plt.plot(ages,py_devs,label="Python Developers",color="#3344FF",linestyle="-.",marker="o",linewidth=3)
plt.plot(ages,js_devs,label="Java Script Developers",color="#FF3355",linestyle=":",marker="x",linewidth=5)
plt.xlabel("Age")
plt.ylabel("Salary in $")
plt.title("Median Salary in $ with Age")
plt.legend()
plt.show()
```

```python
plt.plot(ages,devs,label="All Developers",color="#FF33E9",linestyle="--",marker="D",linewidth=4)
plt.plot(ages,py_devs,label="Python Developers",color="#3344FF",linestyle="-.",marker="o",linewidth=3)
plt.plot(ages,js_devs,label="Java Script Developers",color="#FF3355",linestyle=":",marker="x",linewidth=5)
plt.xlabel("Age")
plt.ylabel("Salary in $")
plt.title("Median Salary in $ with Age")
plt.legend()
plt.tight_layout()
plt.show()
```

Adding Grid to the plot

```python
plt.plot(ages,devs,label="All Developers",color="#FF33E9",linestyle="--",marker="D",linewidth=4)
plt.plot(ages,py_devs,label="Python Developers",color="#3344FF",linestyle="-.",marker="o",linewidth=3)
plt.plot(ages,js_devs,label="Java Script Developers",color="#FF3355",linestyle=":",marker="x",linewidth=5)
plt.xlabel("Age")
plt.ylabel("Salary in $")
plt.title("Median Salary in $ with Age")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
#Resampling the time series data vased on monthly 'M' frequency
df_month = df.resample("M").mean()

#Using subplot
fig, ax = plt.subplots(figsize=(10,6))

#Plotting the bar graph
ax.bar(df_month['2016':].index,df_month.loc['2016':, "Volume"], width=25, align='center')
```

PLOTTING CHANGES IN DATA. IT CAN BE DONE THROUGH MANY WAYS

1.SHIFT : THIS FUNCTION CAN BE SUED TO SHIFT THE DATA BEFORE OR AFTER THE SPECIFIES TIME INTERVAL. WE CAN SEPCIFY THE TIME AND IT WILL SHIFT THE DATA ONE DAY BY DEFAULT.
2.



```python
df['Change']=df.Close.div(df.Close.shift())
df['Change'].plot(figsize=(10,8),fontsize=16)
plt.title("Changing the day by day value")
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

RECOMMENDED TO USE WHEN WE HAVE 5 OR LESS VALUES TO PLOT 

1. PLOTTING THE PIE CHART

```python
slices = [30,40,20,50]
plt.pie(slices)
plt.show()
```

Adding labels to pie chart

```python
labels=['thirty','fourty','twenty','fifty']
plt.pie(slices,labels=labels)
plt.show()
```

Setting the edge color

```python
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'})
plt.show()
```

Setting edge color for slices 

```python
color=['blue','red','yellow','green']
plt.pie(slices,labels=labels,colors=color,wedgeprops={'edgecolor':'black'})
plt.show()
```

Plotting the real world data

```python
labels=['Javascript','HTML/CSS','SQL','Python','Java']
slices=[59219,55466,47544,36443,35971]
```

```python
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'})
plt.show()
```

Exploding the slices

```python
explode=[0,0,0,0.111111111111,0]
plt.pie(slices,labels=labels,colors=color,explode=explode,wedgeprops={'edgecolor':'black'})
plt.show()
```

Broken Pie

```python
explode=[0.2,0.3,0.4,0.111111111111,0.5]
plt.pie(slices,labels=labels,colors=color,explode=explode,shadow=True,wedgeprops={'edgecolor':'black'})
plt.show()
```

```python
explode=[0.2,0.3,0.4,0.111111111111,0.5]
plt.pie(slices,labels=labels,colors=color,explode=explode,startangle=90,wedgeprops={'edgecolor':'black'})
plt.show()
```

```python
explode=[0.2,0.3,0.4,0.1111111111111111111111111111111111111111111111111111111111111111111111,0.5]
plt.pie(slices,labels=labels,colors=color,explode=explode,startangle=90,autopct="%0.1f%%",wedgeprops={'edgecolor':'black'})
plt.show()
```
